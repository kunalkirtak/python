# **Employee Data Analyzer**

```python
from functools import reduce

print("=" * 60)
print("EMPLOYEE DATA ANALYZER")
print("=" * 60)

employees = [
    {"name": "Raj", "department": "IT", "salary": 60000},
    {"name": "Anjali", "department": "HR", "salary": 45000},
    {"name": "Vivek", "department": "Finance", "salary": 70000},
    {"name": "Pooja", "department": "IT", "salary": 90000},
    {"name": "Rohan", "department": "Sales", "salary": 40000},
    {"name": "Aditi", "department": "IT", "salary": 85000},
]

print("\nEmployee Data")

for employee in employees:
    print(employee)

# ------------------------------------------------
# List Comprehension
# ------------------------------------------------

employee_names = [
    employee["name"]
    for employee in employees
]

print("\nEmployee Names")
print(employee_names)

# ------------------------------------------------
# Dictionary Comprehension
# ------------------------------------------------

salary_dictionary = {
    employee["name"]: employee["salary"]
    for employee in employees
}

print("\nSalary Dictionary")
print(salary_dictionary)

# ------------------------------------------------
# map()
# ------------------------------------------------

annual_salary = list(
    map(
        lambda employee: {
            "name": employee["name"],
            "annual_salary": employee["salary"] * 12
        },
        employees
    )
)

print("\nAnnual Salary")

for employee in annual_salary:
    print(employee)

# ------------------------------------------------
# filter()
# ------------------------------------------------

high_salary = list(
    filter(
        lambda employee: employee["salary"] >= 60000,
        employees
    )
)

print("\nEmployees Salary >= 60000")

for employee in high_salary:
    print(employee)

# ------------------------------------------------
# reduce()
# ------------------------------------------------

payroll = reduce(
    lambda total, employee: total + employee["salary"],
    employees,
    0
)

print("\nTotal Monthly Payroll")
print(payroll)

# ------------------------------------------------
# sorted()
# ------------------------------------------------

ranking = sorted(
    employees,
    key=lambda employee: employee["salary"],
    reverse=True
)

print("\nSalary Ranking")

for employee in ranking:
    print(employee["name"], employee["salary"])

# ------------------------------------------------
# Lambda
# ------------------------------------------------

increment = lambda salary: salary * 1.10

print("\nSalary After 10% Increment")

for employee in employees:
    print(
        employee["name"],
        "₹",
        round(increment(employee["salary"]), 2)
    )
```
