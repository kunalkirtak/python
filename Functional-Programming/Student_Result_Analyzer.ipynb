# **Student Result Analyzer**

```python
from functools import reduce

print("=" * 60)
print("STUDENT RESULT ANALYZER")
print("=" * 60)

students = [
    {"name": "Aman", "marks": 92},
    {"name": "Rahul", "marks": 75},
    {"name": "Sneha", "marks": 88},
    {"name": "Priya", "marks": 63},
    {"name": "Vikas", "marks": 48},
    {"name": "Neha", "marks": 95},
]

print("\nOriginal Data")
for student in students:
    print(student)

# ------------------------------------------------
# List Comprehension
# ------------------------------------------------

student_names = [student["name"] for student in students]

print("\nStudent Names")
print(student_names)

# ------------------------------------------------
# Dictionary Comprehension
# ------------------------------------------------

student_marks = {
    student["name"]: student["marks"]
    for student in students
}

print("\nDictionary")
print(student_marks)

# ------------------------------------------------
# map()
# ------------------------------------------------

bonus_marks = list(
    map(
        lambda student: {
            "name": student["name"],
            "marks": student["marks"] + 5
        },
        students
    )
)

print("\nBonus Marks")
print(bonus_marks)

# ------------------------------------------------
# filter()
# ------------------------------------------------

passed_students = list(
    filter(
        lambda student: student["marks"] >= 50,
        students
    )
)

print("\nPassed Students")
for student in passed_students:
    print(student)

# ------------------------------------------------
# reduce()
# ------------------------------------------------

total_marks = reduce(
    lambda total, student: total + student["marks"],
    students,
    0
)

average = total_marks / len(students)

print("\nTotal Marks :", total_marks)
print("Average :", round(average, 2))

# ------------------------------------------------
# sorted()
# ------------------------------------------------

ranking = sorted(
    students,
    key=lambda student: student["marks"],
    reverse=True
)

print("\nRanking")

for i, student in enumerate(ranking, start=1):
    print(i, student["name"], student["marks"])

# ------------------------------------------------
# Lambda
# ------------------------------------------------

grade = lambda marks: (
    "A"
    if marks >= 90
    else "B"
    if marks >= 75
    else "C"
    if marks >= 60
    else "Fail"
)

print("\nGrades")

for student in students:
    print(student["name"], ":", grade(student["marks"]))
```
