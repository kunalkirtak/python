# **Log File Analyzer**

```python
from collections import Counter

# ------------------------------------------------------------
# Create a sample log file
# ------------------------------------------------------------

sample_logs = [
    "INFO User logged in",
    "INFO Dashboard opened",
    "WARNING Disk usage is high",
    "ERROR Database connection failed",
    "INFO File uploaded",
    "ERROR Invalid password",
    "WARNING Low memory",
    "INFO User logged out"
]

log_file = "server_logs.txt"

with open(log_file, "w") as file:
    for log in sample_logs:
        file.write(log + "\n")

# ------------------------------------------------------------
# Generator Function
# ------------------------------------------------------------

def read_logs(filename):
    """Read log file line by line."""

    try:
        with open(filename, "r") as file:
            for line in file:
                yield line.strip()

    except FileNotFoundError:
        print("Log file not found.")

# ------------------------------------------------------------
# Count Log Levels
# ------------------------------------------------------------

counter = Counter()

for log in read_logs(log_file):
    level = log.split()[0]
    counter[level] += 1

print("=" * 50)
print("LOG SUMMARY")
print("=" * 50)

for level, count in counter.items():
    print(f"{level:<10}: {count}")

# ------------------------------------------------------------
# Display ERROR Logs
# ------------------------------------------------------------

print("\nERROR LOGS")

for log in read_logs(log_file):
    if "ERROR" in log:
        print(log)

# ------------------------------------------------------------
# Keyword Search
# ------------------------------------------------------------

keyword = "User"

print(f"\nLogs containing '{keyword}'")

for log in read_logs(log_file):
    if keyword.lower() in log.lower():
        print(log)
```
