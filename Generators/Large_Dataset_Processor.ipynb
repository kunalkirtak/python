# **Large Dataset Processor**

```python
import random

# ------------------------------------------------------------
# Generator Function
# ------------------------------------------------------------

def generate_sales_data(records):

    for record_id in range(1, records + 1):

        yield {
            "ID": record_id,
            "Sales": random.randint(1000, 100000)
        }

# ------------------------------------------------------------
# Process Dataset
# ------------------------------------------------------------

total_sales = 0
high_value_sales = 0
record_count = 0

print("=" * 60)
print("FIRST 10 RECORDS")
print("=" * 60)

for record in generate_sales_data(100000):

    record_count += 1

    total_sales += record["Sales"]

    if record["Sales"] > 80000:
        high_value_sales += 1

    if record_count <= 10:
        print(record)

average_sales = total_sales / record_count

print("\n")
print("=" * 60)
print("DATASET SUMMARY")
print("=" * 60)

print("Total Records :", record_count)
print("Total Sales   :", total_sales)
print("Average Sales :", round(average_sales, 2))
print("High Value Records :", high_value_sales)

# ------------------------------------------------------------
# Generator Expression
# ------------------------------------------------------------

sales_generator = (
    record["Sales"]
    for record in generate_sales_data(100000)
)

maximum_sale = max(sales_generator)

print("\nMaximum Sale Found :", maximum_sale)
```
