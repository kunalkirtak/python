# **Performance Profiler (Timing Decorator)**

```python
import time
import random
from functools import wraps


# ----------------------------------------------------------
# Timing Decorator
# ----------------------------------------------------------

def timing_decorator(func):
    """
    Measures execution time of any function.
    """

    @wraps(func)
    def wrapper(*args, **kwargs):

        start = time.perf_counter()

        result = func(*args, **kwargs)

        end = time.perf_counter()

        execution_time = end - start

        print(f"\nFunction : {func.__name__}")
        print(f"Execution Time : {execution_time:.6f} seconds")

        return result

    return wrapper


# ----------------------------------------------------------
# Example Function 1
# ----------------------------------------------------------

@timing_decorator
def sort_large_list(size):

    numbers = [random.randint(1, 1000000) for _ in range(size)]

    numbers.sort()

    return numbers[:10]


# ----------------------------------------------------------
# Example Function 2
# ----------------------------------------------------------

@timing_decorator
def calculate_sum(limit):

    return sum(range(limit))


# ----------------------------------------------------------
# Example Function 3
# ----------------------------------------------------------

@timing_decorator
def simulate_delay():

    time.sleep(2)

    return "Task Completed"


# ----------------------------------------------------------
# Main Program
# ----------------------------------------------------------

print("=" * 60)
print("PERFORMANCE PROFILER")
print("=" * 60)

top_numbers = sort_large_list(100000)

print("\nTop 10 Numbers")
print(top_numbers)

total = calculate_sum(1000000)

print("\nSum =", total)

message = simulate_delay()

print(message)
```
