# **Application Logger (Logging Decorator)**

```python
import logging
from functools import wraps


# ----------------------------------------------------------
# Configure Logger
# ----------------------------------------------------------

logging.basicConfig(
    filename="application.log",
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s"
)


# ----------------------------------------------------------
# Logging Decorator
# ----------------------------------------------------------

def log_function(func):

    @wraps(func)
    def wrapper(*args, **kwargs):

        logging.info(f"Calling Function : {func.__name__}")

        logging.info(f"Arguments : {args} {kwargs}")

        try:

            result = func(*args, **kwargs)

            logging.info(f"Returned : {result}")

            return result

        except Exception as error:

            logging.exception(error)

            raise

    return wrapper


# ----------------------------------------------------------
# Banking Class
# ----------------------------------------------------------

class BankAccount:

    def __init__(self, owner, balance):

        self.owner = owner

        self.balance = balance


    @log_function
    def deposit(self, amount):

        self.balance += amount

        return self.balance


    @log_function
    def withdraw(self, amount):

        if amount > self.balance:
            raise ValueError("Insufficient Balance")

        self.balance -= amount

        return self.balance


    @log_function
    def check_balance(self):

        return self.balance


# ----------------------------------------------------------
# Main Program
# ----------------------------------------------------------

print("=" * 60)
print("APPLICATION LOGGER")
print("=" * 60)

account = BankAccount("Kunal", 5000)

print("Current Balance :", account.check_balance())

print("After Deposit :", account.deposit(2500))

print("After Withdrawal :", account.withdraw(1000))

try:

    account.withdraw(10000)

except ValueError as error:

    print("Error :", error)

print("\nCheck application.log file to see all recorded logs.")
```
