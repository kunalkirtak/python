# **File Organizer**

```python
from dataclasses import dataclass
from pathlib import Path
import shutil
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(levelname)s - %(message)s"
)

@dataclass
class FileInfo:
    file_name: str
    extension: str


source = Path("sample_files")

source.mkdir(exist_ok=True)

# Create sample files

for file in [
    "resume.pdf",
    "photo.jpg",
    "notes.txt",
    "music.mp3",
    "presentation.pptx"
]:
    (source / file).touch(exist_ok=True)

logging.info("Sample files created")

for file in source.iterdir():

    if file.is_file():

        info = FileInfo(
            file.name,
            file.suffix.replace(".", "")
        )

        destination = source / info.extension.upper()

        destination.mkdir(exist_ok=True)

        shutil.move(
            str(file),
            destination / file.name
        )

        logging.info(
            f"{info.file_name} moved to {destination.name}"
        )

print("Files organized successfully.")
```
