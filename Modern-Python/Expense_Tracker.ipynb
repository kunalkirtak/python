# **Expense Tracker**

```python
from dataclasses import dataclass
from pathlib import Path
from typing import List
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(levelname)s - %(message)s"
)

@dataclass
class Expense:
    category: str
    amount: float


expenses: List[Expense] = [
    Expense("Food", 450),
    Expense("Travel", 300),
    Expense("Shopping", 1800),
    Expense("Food", 250),
    Expense("Bills", 1200)
]

logging.info("Calculating expense summary")

total = sum(expense.amount for expense in expenses)

summary = {}

for expense in expenses:
    summary[expense.category] = (
        summary.get(expense.category, 0)
        + expense.amount
    )

print("=" * 50)
print("EXPENSE REPORT")
print("=" * 50)

for category, amount in summary.items():
    print(f"{category:<12} ₹{amount:.2f}")

print("\nTotal Spending : ₹", total)

report = Path("expense_report.txt")

with report.open("w") as file:

    file.write("Expense Report\n")
    file.write("=" * 30 + "\n")

    for category, amount in summary.items():
        file.write(f"{category}: ₹{amount:.2f}\n")

    file.write(f"\nTotal: ₹{total:.2f}")

logging.info("Expense report saved successfully")
```
