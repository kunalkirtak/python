# **Student Record Manager**

```python
from dataclasses import dataclass
from typing import List
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(levelname)s - %(message)s"
)

@dataclass
class Student:
    name: str
    roll_no: int
    marks: List[int]

    def average(self) -> float:
        return sum(self.marks) / len(self.marks)

    def grade(self) -> str:
        avg = self.average()

        if avg >= 90:
            return "A"

        elif avg >= 75:
            return "B"

        elif avg >= 60:
            return "C"

        return "D"


students = [
    Student("Rahul", 1, [90, 88, 92]),
    Student("Anjali", 2, [75, 70, 80]),
    Student("Priya", 3, [95, 94, 97])
]

logging.info("Displaying student report")

print("=" * 50)
print("STUDENT REPORT")
print("=" * 50)

for student in students:
    print(
        f"{student.name:<10}"
        f" Average: {student.average():.2f}"
        f" Grade: {student.grade()}"
    )
```
